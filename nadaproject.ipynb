```python

data = pd.read_csv('nadaproject.csv')

sns.regplot(x='CGPA', y='coffee per day', data=data, scatter_kws={'alpha':0.7})
plt.title('Relationship between Coffee Consumption per Day and CGPA')
plt.xlabel('Cups of Coffee per Day')
plt.ylabel('CGPA')
plt.grid(True)
plt.show()
```

```python

categories = ['Increases', 'Decreases', 'Remains the same']
counts = [23,1,10]

plt.figure(figsize=(8, 6))
plt.bar(categories, counts, color='skyblue')

plt.title('Coffee Consumption During Exams')
plt.xlabel('Change in Consumption')
plt.ylabel('Number of Students')

plt.show()
```

```python

categories = ['None','1-2', '3-4', '5-6', 'Everyday']
counts = [8,8,2,6,10]

plt.figure(figsize=(8, 6))
plt.bar(categories, counts, color='skyblue')

plt.title('Days per Week of Coffee Consumption')
plt.xlabel('Days per Week')
plt.ylabel('Number of Students')

plt.show()
```

```python


data = pd.read_csv('nadaproject.csv')

cups_per_day = data['coffee per day']

plt.figure(figsize=(8, 6))
plt.boxplot(cups_per_day)

plt.title('Distribution of Cups of Coffee Consumed Per Day')
plt.xlabel('Cups per Day')
plt.ylabel('Frequency')

plt.show()

```
